```python
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')

import copy, os,csv
```

```python
path = "/home/ula/catfish/ca1.csv"
```

```python
positive_percentage = np.array([])
ieg1_percentage = np.array([])
ieg2_percentage = np.array([])
colloc_percentage = np.array([])

with open(path, newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        total_nuc = int(row['total_nuc'])
        ieg1_pos = int(row['2_pos'])
        ieg2_pos = int(row['3_pos'])
        colloc = int(row['colloc'])
        all_pos = ieg1_pos+ieg2_pos - colloc
        positive_percentage = np.append(positive_percentage, all_pos/total_nuc)
        ieg1_percentage = np.append(ieg1_percentage, ieg1_pos/total_nuc)
        ieg2_percentage = np.append(ieg2_percentage, ieg2_pos/total_nuc)
        colloc_percentage = np.append(colloc_percentage, colloc/total_nuc)
        
```

```python
plt.ioff()
plt.hist(positive_percentage)
plt.title("Liczba pozytywnych jÄ…der/ liczba wszystkich jÄ…der")
plt.show()
```

```python
plt.hist(ieg1_percentage)
plt.title("Liczba ieg1 pozytywnych jÄ…der/ liczba pozytywnych jÄ…der")
plt.show()
```

```python
plt.hist(ieg2_percentage)
plt.title("Liczba ieg2 pozytywnych jÄ…der/ liczba pozytywnych jÄ…der")
plt.show()
```

```python
plt.hist(colloc_percentage)
plt.title("Liczba jÄ…der zawierajÄ…cych oba iegi/ liczba pozytywnych jÄ…der")
plt.show()
```
