```python
import matplotlib.pyplot as plt
%matplotlib notebook
import matplotlib.patches as mpatches
import numpy as np
import copy
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

ellipse = mpatches.Ellipse((3,4), 1, 2, 0,fill=False, edgecolor='red', linewidth=1) 
el2 = copy.copy(ellipse)
print(ellipse.contains_point((3,4)))
ellipse.set_color('green')
ax.add_patch(ellipse)  
ax.set_xlim(5)  
ax.set_ylim(5)
fig.tight_layout()
plt.show()
```
