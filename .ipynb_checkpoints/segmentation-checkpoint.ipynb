```python
import mahotas as mh
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from IPython.html.widgets import interact, fixed
from skimage import io, color
```

```python
dapi_path = "/media/ula/D/adam_catfish/catfish_dapi_tst.gif"
dapi_img = io.imread(dapi_path)
plt.imshow(dapi_img)
```

```python
from skimage.morphology import opening, erosion, disk, closing

selem = disk(3)
eroded = erosion(dapi_img, selem)

tst = opening(dapi_img, selem)
plt.imshow(tst)
```

```python
#dapi_f = mh.gaussian_filter(tst, 2.)
T_mean = tst.mean()
mask = tst < T_mean
plt.imshow(mask)
```

```python
masked = np.ma.masked_array(data = tst,
             mask = mask,fill_value = 0)
filled = masked.filled()


plt.imshow(filled)
```

```python
from skimage.filters import sobel
from skimage.feature import canny

elevation_map = sobel(filled)

edges = canny(elevation_map, sigma=0.2)


plt.imshow(closing(edges))
```

```python
markers = np.zeros_like(dapi_img)
markers[dapi_img < 30] = 1
markers[dapi_img > 150] = 2

plt.imshow(markers)
```

```python
from skimage import morphology
from scipy import ndimage as ndi

from skimage.feature import peak_local_max

distance = ndi.distance_transform_edt(elevation_map)
local_maxi = peak_local_max(distance, indices=False, footprint=np.ones((12, 12)),
                            labels=distance)
markers = ndi.label(local_maxi)[0]






segmentation = morphology.watershed(elevation_map, markers)

plt.imshow(segmentation)
```

```python
from skimage import data
from skimage.morphology import disk
from skimage.filters import threshold_otsu, rank
from skimage.util import img_as_ubyte

radius = 5
selem = disk(radius)

local_otsu = rank.otsu(elevation_map, selem)
plt.imshow(dapi_img >= local_otsu)
```
